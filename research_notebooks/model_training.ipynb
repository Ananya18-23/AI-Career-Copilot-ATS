#### import libraries

```python
import pandas as pd
import numpy as np
```

```python
# 1. Ek realistic sample data ka dictionary banate hain
data = {
    'Skills': [
        'Python, SQL, Machine Learning, Data Visualization, Statistics, Pandas',
        'HTML, CSS, JavaScript, React, Node.js, Web Development, Git',
        'Java, Spring Boot, Microservices, SQL, Docker, AWS, Git',
        'Python, SQL, Tableau, Excel, PowerBI, Data Analysis, Communication',
        'Linux, AWS, Docker, Kubernetes, CI/CD, Jenkins, Shell Scripting',
        'Python, Deep Learning, NLP, TensorFlow, Computer Vision, PyTorch',
        'SQL, Excel, Data Warehousing, ETL, Python, Dashboarding',
        'HTML, CSS, Bootstrap, JavaScript, Vue.js, UI/UX Design',
        'Java, Kotlin, Android Studio, Mobile App Development, Firebase',
        'Cybersecurity, Networking, Ethical Hacking, Firewall, Linux, Wireshark'
    ] * 50, # Isse 500 rows ka ek achha dataset ban jayega
    
    'Role': [
        'Data Scientist',
        'Web Developer',
        'Backend Engineer',
        'Data Analyst',
        'DevOps Engineer',
        'AI/ML Engineer',
        'Data Engineer',
        'Frontend Developer',
        'Mobile App Developer',
        'Cybersecurity Analyst'
    ] * 50
}
```

```python
# 2. DataFrame 
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# 3. Data ko thoda random mix (shuffle) kar dete hain taaki training acchi ho
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# 4. Ise 'data' folder ke andar CSV file banakar save karna
import os
os.makedirs('data', exist_ok=True)
df.to_csv('data/career_data.csv', index=False)
```
